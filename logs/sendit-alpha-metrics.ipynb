# Sendit (Alpha) Metrics

This is the first round of sendit, and we want to look at files (or MB) processed per second, and per minute.

```python
import pandas
from glob import glob
glob('*.tsv')
```

```python
files = glob('*.tsv')
df = pandas.read_csv(files[0],sep="\t",index_col=0)
done = df[df.status=="DONE"]
print("Folders that are done: %s" %done.shape[0])
done
```

Above we see the data frame of data collected, filtered to those that have a status of DONE. This means we can be confident about a start and finish time being present. 

## Plotting Distributions of MB per Time Unit
Below we will plot first MB per minute, and then MB per second. Notice that there is a slight bimodal quality to the plot, likely because we switched from 1 core to 16 (each core corresponding to a worker) sometime in the middle. Also notice the long tail to the right - we had a few bugs that would stop processing, and artificially increase the time. First, we can look at MB per minute.

```python
%matplotlib inline
import seaborn as sns
sns.set()
mb_min = done.size_mb / done.total_time_min
ax = sns.distplot(mb_min)
ax.set_xlabel("MB per minute, N=%s folders" %done.shape[0])
```

```python
mb_sec = done.size_mb / done.total_time_sec
ax = sns.distplot(mb_sec)
ax.set_xlabel("MB per second, N=%s folders" %done.shape[0])
```

```python
import seaborn as sns; sns.set(color_codes=True)
ax = sns.regplot(x=done.total_time_min, y=done.size_mb)
ax.set_xlabel("total time minutes")
ax.set_ylabel("total size (MB)")
```

And here we see a scatterplot of time (minutes) on the X axis and total size (MB) on the Y axis. Again, notice how there is a clump around ~1000 MB / 10 minutes, but then quite a bit of fluctuation. It will be good to run a (non stopping) pipeline, with certainty of all cores, to see more consistent metrics. Now, let's quickly glance at acutal means and standard deviations:

```python
done.mean()
```

```python
done.std()
```

Finally, let's (for the purposes of looking closer at the distribution) crop out the outliers and look to see if there are two means (that possibly represent the two states of 1 and 16 cores)

```python
mb_sec = done.size_mb / done.total_time_sec
ax = sns.distplot(mb_sec)
ax.set_xlabel("MB/sec cropping outliers, N=%s folders" %done.shape[0])
ax.set_xlim(0,6)
```
